```python
from rouge import FilesRouge

sumCandidato = '/content/drive/MyDrive/Proj_PFC/ROUGE/Sumarios_Teste/MMR/Hotel5_MMR.txt'
sumReferencia = '/content/drive/MyDrive/Proj_PFC/ROUGE/Sumarios_Gold/Anotador4/Hotel5_Geral_Anotador4.txt'
resultadosArq = open('/content/drive/MyDrive/Proj_PFC/ROUGE/Resultados/Anotador4/MMR/Hotel5_MMR.txt', 'w')

rouge = FilesRouge()
resultados = rouge.get_scores(sumCandidato, sumReferencia, avg=True, ignore_empty=True)


# Pega o nome dos sumários atuais
nomeSumarioCand = sumCandidato.split('/')
nomeSumarioCand = nomeSumarioCand[len(nomeSumarioCand)-1]
nomeSumarioRef  = sumReferencia.split('/')
nomeSumarioRef  = nomeSumarioRef[len(nomeSumarioRef)-1]

resultadosArq.write("Sumario Candidato: " + nomeSumarioCand + " | Sumário Referência: " + nomeSumarioRef)
resultadosArq.write('\n')
resultadosArq.write('\n')

for key,value in resultados.items():
  print(str(key) + ':' + str(value))
  resultadosArq.write(str(key) + ':' + str(value))
  resultadosArq.write('\n')
```
